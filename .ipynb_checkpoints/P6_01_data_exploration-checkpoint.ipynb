```python
# Fonction pour converture le contenu du champs product_specifications
def json_reformat(a):
    a="{"+a[26:-1].replace('\"key\"=>',"").replace('\"value\"=>',"").replace("\", \"","\":\"").replace("[","").replace("]","").replace("{","").replace("}","")+"}"
    return a

description["product_specifications2"]=description["product_specifications"].apply(lambda x: json_reformat(x) if x is not np.nan else "")
```

```python
# Reformattage de l'arborescence de la categorisation du produit
description["product_category_tree2"]=description["product_category_tree"].apply(lambda x: x[2:-2])
description["product_category_tree3"]=description["product_category_tree2"].apply(lambda x: x.split(" >> "))
description["product_category_tree3"]
```

```python
description["product_category_tree4"]=description["product_category_tree3"].apply(lambda x: " ".join(x[:2]))
```

# Image transformation

```python
# Transformations des donnees
transform = A.Compose([
    A.HorizontalFlip(p=0.5),
    A.RandomBrightnessContrast(p=0.2),
])
```

```python
# Data augmentation avec changement de contrast apres uniformisation de l'historigramme des valeurs de pixel et floutage
for a in description["image"]:
    im1 = Image.open("../data/Dataset+projet+prétraitement+textes+images/Flipkart/Images/"+a)
    test = ImageOps.autocontrast(im1)
    test = ImageOps.equalize(test)
    test = test.filter(ImageFilter.BoxBlur(1))
    test = np.array(test)
    test = transform(image=test)["image"]
    test = Image.fromarray(np.uint8(test))
    test.save("../data/data_pretraitement/data1/"+a)
```
